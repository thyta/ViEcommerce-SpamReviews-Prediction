# pie chart

```python
from math import pi
from bokeh.palettes import Category20c, Category20
from bokeh.plotting import figure
from bokeh.transform import cumsum
pn.extension()

import matplotlib.pyplot as plt
```

```python
#create arry for pie chart
rate_df = df['Rating']
rate_df = rate_df.sort_values()
rate_df
```

```python
rate_array = rate_df.to_numpy()
print(rate_array)
```

```python
rate_array.shape
```

```python
# rate_count_01 = {i:rate_array.count(i) for i in rate_array}
# print(rate_count_01)
from collections import Counter

rate_count_01 = Counter(rate_array)

rate_count_01
```

```python
data_rate = pd.Series(rate_count_01).reset_index(name='value').rename(columns={'index':'rating'})
data_rate['angle'] = data_rate['value']/data_rate['value'].sum() * 2*pi
data_rate['color'] = Category20c[len(rate_count_01)]
```

```python
p_rate = figure(height=350, title="Pie Chart Rating", toolbar_location=None,
           tools="hover", tooltips="@rating start rate: @value", x_range=(-0.5, 1.0))
```

```python
r_rate = p_rate.wedge(x=0, y=1, radius=0.4,
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='rating', source=data_rate)
```

```python
p_rate.axis.axis_label=None
p_rate.axis.visible=False
p_rate.grid.grid_line_color = None
```

```python
bokeh_pane = pn.pane.Bokeh(p_rate, theme="dark_minimal")
```

```python
# Assuming the DataFrame is named 'df'
rating_counts = df['Rating'].value_counts().sort_index()

# Create a Markdown pane with the rating counts
rating_list = '\n'.join([f"Rating {rating}: {count} occurrences\n" for rating, count in rating_counts.items()])
rating_pane = pn.pane.Markdown(f"## Rating Counts\n{rating_list}")
rating_pane
```

```python

# Filter the dataframe to include only rows with a rating of 5
df_5_star = df[df['Rating'] == 5]

# Group the data by product category and calculate the count of 5-star ratings
category_ratings = df_5_star.groupby('Product Categories')['Rating'].count()

# Get the product category with the highest count of 5-star ratings
best_category = category_ratings.idxmax()

# Get the count of 5-star ratings for the best category
best_category_count = category_ratings.max()

# Create a panel widget to display the best product category and its count
best_category_widget = pn.pane.Str("Best Product Category:")
best_category_widget.object += f"\nCategory: {best_category}"
best_category_widget.object += f"\nCount of 5-Star Ratings: {best_category_count}"
```

```python
# Filter the dataframe to include only rows with spam reviews (Class == 1)
spam_reviews = df[df['Class'] == 1]

# Group the data by product and calculate the count of spam reviews
spam_counts = spam_reviews.groupby('Product Categories')['Class'].count()

# Get the product with the most spam reviews
product_with_most_spam = spam_counts.idxmax()

# Get the count of spam reviews for the product with the most spam
most_spam_count = spam_counts.max()

# Get the product with the least spam reviews
product_with_least_spam = spam_counts.idxmin()

# Get the count of spam reviews for the product with the least spam
least_spam_count = spam_counts.min()

# Create widgets for the sidebar
most_spam_widget = pn.widgets.StaticText(
    name="Product with the most spam reviews:",
    value=product_with_most_spam
)

most_spam_count_widget = pn.widgets.StaticText(
    name="Number of spam reviews for the product:",
    value=str(most_spam_count)
)

least_spam_widget = pn.widgets.StaticText(
    name="Product with the least spam reviews:",
    value=product_with_least_spam
)

least_spam_count_widget = pn.widgets.StaticText(
    name="Number of spam reviews for the product:",
    value=str(least_spam_count)
)
```

# Teamplate 

```python
# Create the FastListTemplate
template = pn.template.FastListTemplate(
    title='Report dashboard for products in shopee platforms',
    sidebar=[
        rating_pane,
        best_category_widget,
        most_spam_widget,
        most_spam_count_widget,
        least_spam_widget,
        least_spam_count_widget
    ],
    main=[
        pn.Row(pn.Column(bokeh_pane)),
        pn.Row(pn.Column(bar_chart_normal)),
        pn.Row(pn.Column(bar_chart_spam)),
    ],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)
```
